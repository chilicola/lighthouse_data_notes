```python
from nltk.corpus import comtrans
```

```python
# load words
words = comtrans.words('alignment-en-fr.txt')
for word in words[:20]:
    print(word)
```

```python
print(words[0])
```

```python
# load sentences
als = comtrans.aligned_sents('alignment-en-fr.txt')[0]
als
```

```python
print(' '.join(als.words))
print(' '.join(als.mots))
```

```python
# alignments
als.alignment
```

```python
# display alignment
als.invert()
```

```python
# create alignments
from nltk.translate import Alignment, AlignedSent
```

```python
als = AlignedSent(['Reprise', 'de', 'la', 'session'], ['Resumption', 'of', 'the', 'session'], Alignment([(0,0), (1,1), (2,2), (3,3)]))
als
```

### Translating with IBM Model 1 in NLTK

```python
from nltk.translate import IBMModel1
```

```python
com_ibm1 = IBMModel1(comtrans.aligned_sents()[:10], 100)
com_ibm1
```

```python
print(round(com_ibm1.translation_table['bitte']['Please'], 3))
```

```python
print(round(com_ibm1.translation_table['Stizungsperiode']['session'], 3))
```
