## Matplotlib 3D Assignment

```python
# import mplot3d
from mpl_toolkits import mplot3d

# import numpy
import numpy as np

# import matplotlib
import matplotlib.pyplot as plt
```

------------

#### Task 1: Create wireframe of hyperbolic paraboloid on the domain <-10,10>X<-10,10>

* hyperbolic paraboloid equation is: z = x*x - y*y

```python
# function for z

def f(x, y):
    return(x**2 - y**2)
```

```python
# evaluate f() or z, on 100x100 uniformly spaced samples in [-10, 10]x[-10, 10]

t = np.linspace(-10,10,500)
x, y = np.meshgrid(t, t)

z = f(x, y)
```

```python
# visualize

fig = plt.figure(figsize = (12.8, 9.6))
ax = plt.axes(projection = '3d')

ax.plot_wireframe(x, y, z)
```

#### Task 2: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* eliptical paraboloid : z = x*x + y*y

```python
# function for z

def f(x, y):
    return(x**2 + y**2)
```

```python
# define coordinations

t = np.linspace(-10, 10, 500)
x, y = np.meshgrid(t, t)

z = f(x, y)
```

```python
# plot 

# canvas
fig = plt.figure(figsize=(12.8, 9.6))

# axes
ax = plt.axes(projection = '3d')

ax.plot_surface(x, y, z)
```

#### Task 3: Create surface plot of eliptical paraboloid on the domain <-10,10>X<-10,10>

* Also create a contour projection on the plane z=0 (HINT: use the parameter 'zdir' of contour plot)

* eliptical paraboloid : z = x*x + y*y

```python
# function for z

def f(x, y):
    return(x**2 + y**2)
```

```python
# set coordinates

t = np.linspace(-10, 10, 500)
x, y = np.meshgrid(t, t)
z = f(x, y)
```

```python
# plot

fig = plt.figure(figsize = (12.8, 9.6))
ax = plt.axes(projection = '3d')
ax.contour(x, y, z, zdir = 'z')
```
