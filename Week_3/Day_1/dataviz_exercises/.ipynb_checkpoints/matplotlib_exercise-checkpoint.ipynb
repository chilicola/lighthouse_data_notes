## Matplotlib Assignment

```python
# import pandas
import pandas as pd

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('company_sales_data.csv')
```

```python
df.head()
```

### Task 1: Read the `total_profit` of all months and display it using a line plot.

```python
x = df['month_number']
y = df['total_profit']

plt.plot(x, y)
```

#### Stateless Approach

```python
# set axis

x = df['month_number']
y = df['total_profit']
```

```python
# set plot

fig = plt.plot(x, y, 'red')
```

### Task 2: Read all the different product sales data and display it using a multiline plot.

```python
x = df['month_number']

y1 = df['facecream']
y2 = df['facewash']
y3 = df['toothpaste']
y4 = df['bathingsoap']
y5 = df['shampoo']
y6 = df['moisturizer']

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
```

#### Stateless Approach

```python
# set axis

x = df['month_number']

y1 = df['facecream']
y2 = df['facewash']
y3 = df['toothpaste']
y4 = df['bathingsoap']
y5 = df['shampoo']
y6 = df['moisturizer']
```

```python
# set plot

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)
ax.plot(x, y4)
ax.plot(x, y5)
ax.plot(x, y6)
```

### Task 3: Read the `total_profit` of each month and display it using a histogram to find out which profit ranges are the most common.

```python
x = df['total_profit']

plt.hist(x)
plt.xlabel('Total Profit')
plt.ylabel('Counts')
plt.title('Profit by Counts')
plt.show()
```

```python
# for practice only

x = [df['month_number'][i] for i in range(len(df['month_number']))]
y = [df['total_profit'][i] for i in range(len(df['total_profit']))]

plt.bar(x, y)
```

#### Stateless Approach
