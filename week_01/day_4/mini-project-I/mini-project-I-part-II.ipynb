## Task
Search for you favorite movie and return the information about the movie:
- name
- release date
- runtime
- revenue
- length of the overview
- number of votes
- voting average

```python
param_get_moive = f'?query="lord of the rings"&api_key={TMDB_KEY_V3}'
res_moive = requests.get(base + get_movie + param_get_moive)

get_movie_detail = f'/movie/121'
param_get_movie_detail = f'?api_key={TMDB_KEY_V3}'
res_movie_detail = requests.get(base + get_movie_detail + param_get_movie_detail)
```

```python
res_moive.json()
```

```python
res_detail.json()
```

```python
name = res_moive.json()['results'][0]['original_title']
release_date = res_moive.json()['results'][0]['release_date']
runtime = res_movie_detail.json()['runtime']
revenue = res_movie_detail.json()['revenue']
len_of_overview = len(res_movie_detail.json()['overview'])
vote_count = res_moive.json()['results'][0]['vote_count']
vote_avg = res_moive.json()['results'][0]['vote_average']

print(f'Name: {name}')
print(f'Release Date: {release_date}')
print(f'Runtime: {runtime}')
print(f'Revenue: {revenue}')
print(f'Length of Overview: {len_of_overview}')
print(f'Vote Count: {vote_count}')
print(f'Vote Average: {vote_avg}') 
```

## Task 
Find top 5 trending:
- movies
- tv-shows
- people

```python
get_movie_trend = '/trending/movies/day'
params_get_movie_trend = f'?api_key={TMDB_KEY_V3}'
res_get_movie_trend = requests.get(base + get_movie_trend + params_get_movie_trend)

get_tv_trend = '/trending/tv/day'
params_get_tv_trend = f'?api_key={TMDB_KEY_V3}'
res_get_tv_trend = requests.get(base + get_tv_trend + params_get_tv_trend)

get_person_trend = '/trending/person/day'
params_get_person_trend = f'?api_key={TMDB_KEY_V3}'
res_get_person_trend = requests.get(base + get_person_trend + params_get_person_trend)
```

```python
res_get_movie_trend.json()
```

```python
res_get_tv_trend.json()
```

```python
trend_movie = []
for res in res_get_movie_trend.json()['results'][:5]:
    if 'original_title' in res:
        trend_movie.append(res['original_title'])
    else:
        trend_movie.append(res['original_name'])

trend_tv = [tv['name'] for tv in res_get_tv_trend.json()['results'][:5]]

trend_person = [person['name'] for person in res_get_person_trend.json()['results'][:5]]

print(f'Trending Movies: {trend_movie}')
print(f'Trending TVs: {trend_tv}')
print(f'Trending People: {trend_person}')
```
