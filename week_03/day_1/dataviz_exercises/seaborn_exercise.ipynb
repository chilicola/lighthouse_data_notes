## Seaborn Assignment

```python
# import pandas
import pandas as pd

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv',sep=',')
df.drop('#',axis=1,inplace=True)
```

```python
df.head()
```

### Task 1: Plot a regression line between attack and defense.

```python
sns.lmplot(x='Attack', y='Defense', data = df)
plt.show()
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
sns.lmplot(x='Attack',y='Defense', data = df, hue = 'Generation')
plt.show()
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
plt.subplot(2,3,1)
sns.boxplot(x='Type 1',y='HP', data = df)

plt.subplot(2,3,2)
sns.boxplot(x='Type 1',y='Attack', data = df)

plt.subplot(2,3,3)
sns.boxplot(x='Type 1',y='Defense', data = df)

plt.subplot(2,3,4)
sns.boxplot(x='Type 1',y='Sp. Atk', data = df)

plt.subplot(2,3,5)
sns.boxplot(x='Type 1',y='Sp. Def', data = df)

plt.subplot(2,3,6)
sns.boxplot(x='Type 1',y='Speed', data = df)

plt.show()
```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
df_v1 = df.drop(columns = ['Generation', 'Legendary'])

# compute correlations between features
v1_corr = df_v1.corr()

# plot the correlations
sns.heatmap(v1_corr)
plt.title('Correlation plot')
plt.show()
```
