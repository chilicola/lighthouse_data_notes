```python
# import pandas
import pandas as pd  

# from apyori import apriori
from apyori import apriori
```

```python
# load data 
movie_data = pd.read_csv('https://raw.githubusercontent.com/pirandello/apriori/master/movie_dataset.csv', 
                         header = None)
```

* transform dataframe to list of lists (suitable format for apyori)

```python
range(len(movie_data))
```

```python
records = []
for i in range(len(movie_data)):
    records.append([str(movie_data.values[i, j]) for j in range(len(movie_data.columns))])
```

* instantiate apriori and set the input params based on following constraints:
    * we want only movies that are purchased at least 40 times
    * the minimum confidence for the rules is 20%
    * the minumum lift is 3

```python
association_rules = apriori(records, min_support=40/7501, min_confidence=0.2, min_lift=3)
association_rules = list(association_rules)
```

* how many association rules did we obtained ?

```python
len(association_rules)
```

* print the first association rule

```python
association_rules[0]
```

* convert association rules to DataFrame 
    * use these columns: title_1, title_2, support, confidence, lift

```python
# initialize features
names_1 = []
names_2 = []
supports = []
confidences = []
lifts = []

# extract all features
for items in association_rules:
    # extract items names
    names = [item for item in items[0]]
    names_1.append(names[0])
    names_2.append(names[1])
    
    # extract support
    supports.append(items[1])
    
    # extract confidence
    confidences.append(items[2][0][2])
    
    # extract lift
    lifts.append(items[2][0][3])
```

```python
# create dataframe
rules_dict = {'title_1': name_1, 'title_2': name_2, 'support': supports, 'confidence': confidences, 'lift':lifts}
df = pd.DataFrame(rules_dict)
```
